# Regression vs. Classification: What's the difference?

So, we have covered that machine learing is for making predictions about the future based on data collected. These tyeps of systems are known as Predictive modleing. They function by takeing data in(x) and generating a probible outcome(y). 
There are two types of Preditive modleing, Regression and Classification. Selecting which type works for the job you are trying to accomplish is the first step in working to solve your problem. So what is a Regression and Classification algorithm and what are they used for?

Regression is used to predict continuous numerical values, such as the price of a stock or the age of a person. In a regression problem, the goal is to fit a model to the training data, so that when provided with new, unseen data, the model is able to make accurate predictions about the continuous target variable. For example, linear regression and polynomial regression are both regression.Regression predictions can be evaluated using root mean squared error, whereas classification predictions cannot.

Classification, on the other hand, is used to predict categorical values, such as whether a customer will buy a product or not, or whether an email is spam or not. The goal of a classification problem is to learn a mapping from the input variables (features) to one of several possible output classes. For example, logistic regression, k-nearest neighbors, and support vector machines are all examples of classification. Classification predictions can be evaluated using accuracy, whereas regression predictions cannot.

#The most basic machine learning algorithm: Linear Regression

Linear regression is a regression algorithm used to predict an outcome based on the interaction between two varibles. These variables are divided into two groups: independent and dependent. You can think of independent variables like your inputs and the dependent variables as your outputs. Dependent variables are controlled by the independent variables. The idea behind linear regression is that you assume there is an approximately linear relationship between the inputs and outputs so that we can look for some line that describes this relationship to make predictions about the dependent variable, given some new independent variables. f(x)= mx+b 

 where Y is the dependent variable, and the values m and b are produced by linear regression. By useing this modle a line of best fit is created that can be used to dettermine the most likely result givein a value.

# Logistic Regression: Just Linear Regression with Extra Steps

Even though it doesn't sound like it, logistic regression is a classification algorithm, which is used to predict a binary outcome based on a set of independent variables. Take the example we used above: hours of sleep vs. a student's test grade. Now, let's change the problem a little. Consider the case where we want to determine if the student will pass the test based on the hours of sleep. This is a simple yes or no question-- you don't need it to output any continuous values like we did in linear regression, we just want a simple yes/no. How do we do this?  

In logistic regression, we assume that the relationship between the dependent variable and independent variables is **linearly separable**. This means that, if you look at the grouping of all the data points, you can drop a line somewhere in between them to separate the groups. Notice the nuance here; this is one of the differences between logistic and linear regression. In linear regression, we assume the data can be **modeled linearly**. That means that we assume that there is some line that fits to (touches) the data points. Logistic regression, however, assumes that the data is **linearly separable**. That is, you can separate the data into discrete classes. The below illustration should make it more clear:

The goal is to find the best fitting line (also known as the regression line) that minimizes the difference between the observed values and the values predicted by the line. The line is represented by the equation of a straight line, where the coefficients represent the relationship between the independent and dependent variables.

But how does the system determine what group it is in? This is done by useing a mathmatical equation to put it in one of two groups. Now, we are going to present this equation slightly differently than you may see it presented online. We are going to write it as a composite of two different functions. 

First, $ F(x) = -(mx+b) $.

Next, $G(x) = \frac{1}{1+e^{x}}$

Then, composing them, 

$G(F(x))=\frac{1}{1+e^{-(mx+b)}}$.

Let's dive into what these functions actually do. 

Notice the power to which the exponential is raised: $ -(mx+b) $. That's our F(x), and it's also the slope intercept form of the equation of a line -- the exact same thing that powers linear regression, just put within another equation! Logistic Regresion sounds rather intimidating, but really, it's just linear regression (f(x)=mx+b) squished inside a new function! And linear regression, as we already covered, is basically just the slope-intercept form of an equation of a line; we just use the computer to find m and b!

# What's a Perceptron? 

A Perceptron is a type of artificial neural network that is capable of binary classification. It was invented by Frank Rosenblatt in 1957 and is one of the oldest and simplest types of neural networks.

> The Perceptron consists of a single layer of nodes, where each node represents a feature or input variable. These input variables are multiplied by their respective weights and then summed together. The sum is then passed through an activation function, which determines the output of the node. The output of each node is either 0 or 1, which represents the binary classification of the input data.

## Mathematical Representation: Essentially Linear Regression

Neural networks are a class of machine learning models that are inspired by the structure and function of the human brain. They are composed of multiple interconnected processing units, called neurons, that work together to perform complex computations.

At their core, neural networks rely heavily on linear regression, which is a simple and powerful statistical method for modeling the relationship between two variables. Linear regression assumes that there is a linear relationship between the independent variable (or variables) and the dependent variable, and tries to find the best line (or plane, or hyperplane) that fits the data.

In a neural network, each neuron can be thought of as a linear regression model. It takes in one or more inputs, multiplies each input by a weight, and adds them up to produce an output. This output is then passed through an activation function, which introduces nonlinearity into the model.

During the training process, the neural network adjusts the weights and biases of each neuron to minimize the difference between the predicted outputs and the actual outputs. This process is typically done using a method called backpropagation, which calculates the gradient of the loss function with respect to the weights and biases, and updates them accordingly.

In summary, neural networks essentially "use" linear regression by breaking down complex computations into a series of simpler linear operations, and introducing nonlinearity through activation functions. By doing so, they can learn to model complex relationships between inputs and outputs, and make accurate predictions on new data

## The Activation Function: Essentially Logistic Regression

The activation function is a crucial component of a neural network, which introduces nonlinearity into the model. It takes in the weighted sum of the inputs to a neuron and produces an output, which is then passed on to the next layer of the network.

One commonly used activation function is the logistic function, also known as the sigmoid function. The sigmoid function maps any input to a value between 0 and 1, which can be interpreted as a probability. It has the following formula:

f(x) = 1 / (1 + e^-x)

This function is essentially logistic regression, which is a statistical method for modeling the probability of a binary outcome based on one or more predictor variables.

In logistic regression, the model calculates the probability of the outcome variable being 1 (or "yes") given the predictor variables. The logistic function is then used to transform this probability into a value between 0 and 1, which can be interpreted as the predicted probability of the outcome.

In a neural network, the sigmoid function is used as the activation function to introduce nonlinearity into the model. The weighted sum of the inputs to a neuron is first calculated, and then the sigmoid function is applied to this sum to produce the output. This output is then passed on to the next layer of the network, where the process is repeated.

During the training process, the neural network adjusts the weights and biases of each neuron to minimize the difference between the predicted outputs and the actual outputs. This process is typically done using a method called backpropagation, which calculates the gradient of the loss function with respect to the weights and biases, and updates them accordingly.

In summary, the activation function, particularly the sigmoid function, can be seen as essentially logistic regression in machine learning, as they both transform a weighted sum of inputs into a probability or value between 0 and 1.

## Let's Walk Through it

Okay, you have been introduced to a multilayer perceptron -- the building block of all modern neural networks and the foundation of deep learning. Let's now take a moment to explore the three steps that take place in a neural network at a conceptual level. Don't worry, we aren't going to do a deep math dive here; we just want you to understand the intuition behind how a neural network learns. Essentially, it boils down to three stages that are pretty similar to how we learn. 

Let's think about in terms of a suprevised classification problem: let's say we have a dataset generated by a company regarding employee performance (we are going to actually use a dataset like this later, although it will have some differences). The dataset records three values: employee age, a performance rating, and whether the employee recieved a promotion. We can train a neural network to predict whether the employee recieved a promotion by giving it the age and performance score data as input and using the "promotion" column as a label for our data. The first stage in our training is what we call the forward pass. 

### The Forward Pass

The forward pass, in short, is where the network makes its guess. Under the hood, we are applying all our weights and biases -- all our linear regressions -- to the input data and we get some number as an output. At first, we aren't generating anything useful because we haven't found "good" values for all "m" and "b" terms in each neuron. The process of finding those good values constitute the other two steps in neural network learning. 

### The Cost Function

The first step in finding good m and b values is figuring out how incorrect our initial guess is. This process is called a cost function, or a loss function. The best way to understand the intuition behind this is through an example, so let's go back to our employee promotion dataset example. Suppose we have an employee who is 52 with a performance rating of 8/10 who was promoted. Let's assume that, numerically, people who were promoted have a value "1" in the promoted column and those that weren't have a "0" in that column. Remember that, at the start of training, all our "m's" and "b's" are just random numbers, so our first forward pass is going to be nonsense. 

Let's define our loss function to be $ f = (correct Value - predicted Value) $. 
Suppose that our first forward pass yields " .2. The loss function finds how wrong we are. Plugging in, we get that $ f = 1 - .2 = .8 $.  

For our example here, we used loss function $ f = (correct Value - predicted Value) $. However, in practice, machine learning users typically employ more complicated cost functions. However, there is one thing they all have in common -- they are all just different ways to measure how wrong your model's guess (its forward pass) was. 

A quick confession: technically, we lied to you in the earlier portion of this section. We said that the loss function and cost function are the same, but there is actually a nuanced difference between the two. Technically, the loss function is **only for a single datapoint** whereas the cost function **is the average loss function over the entire training dataset**. 

So, let's recap. We discussed that a neuron in a neural network is a composable function G(F(X)) where F(X) is essentially just linear regression and G(X) (called the activation function) serves the same purpose that the softmax function does in logistic regression: to introduce nonlinearity. To make a deep neural network (at least a basic one) we stack layers of these neurons on top of each other. We have discussed the forward pass, which is how the neural network guesses a value for some given input. We discussed that this is just the output of all the little linear and logistic regressions in each neuron of the network. Then, we discussed the cost/loss function, which is a mechanism for us to figire out how wrong our network's guess was. Now, we will discuss the most important part of the neural network: backpropogation. 

### Backpropogation

Backpropogation is the power of the neural network. The Backpropogation algorithm is what lets our network learn from its mistakes. The general structure is this: 

1. the cost function outputs the loss from a training batch
2. the algorithm then calulates what changes to the weights and biases in the last layer would have moved us closer to the correct answer
3. then we repeat step 2 for the next layer in the neural network until we are out of layers. 

Let's walk through it conceptually with an example. Imagine you are a neural network, and let's say I give you a huge stack of images of cats and dogs. Since you are a neural network (a computer), you only work on numbers. So I tell you our label for cats = 0 and dogs = 1. Then, I show you an image and tell you to guess 0 or 1 (cat or dog). You guess 4.2. 

Why 4.2? Because you are a new neural network and you know nothing -- that's just what your randomly initialized weights and biases produced. This is your first forward pass, and you need time to learn. So, I run our cost function. I was hoping you would say 1 since I showed you a cat, but you said 4.2. That means you were off by 3.2 at the last layer. Again, the cost functions we use in real-life are more complex than the simple subtraction we are using here, but they are still just measures of how wrong we were. **Remember that a neural network is deep -- that is, a neural network is a series of perceptrons stacked into layers, where each layer has associated with it weights and biases to produce the next layer**. Pay attention to this next part: this is the crux of backpropogation. We take our error (3.2) and ask ourselves *what values for weights and biases at the previous layer would have produced 3.2 at the output*. Then, we update the weights and biases to what we calulate. 

Okay, so now your last layer has updated its weights and biases to give us the output we want. But, what about the layer before that? **Remember that the weights and biases at each layer affect the next**. We need to repeat this process. We ask "What values for the previous layer's weights and biases *would have* given us the weights and biases we wanted at the last layer?". We ask this question at every layer, starting at the end and working our way back through the network until we have updated all the weights and biases (all our m's and b's) to give us the output we wanted. 

If you understand that, you understand the essence of backpropogation, and how all neural networks learn. But, yet again, we must make a slight confession -- we lied again. After we calculate our loss, Backpropogation does not technically change our weights to be *exactly* what would have given us the correct answer. Instead, it moves the weights and biases **a small amount** in the correct direction toward being right. 

Thais probably feels a bit weird. After all, why wouldn't we want our neural network to be completely right about our cat image? But, stop and think about what it would mean for us to update our weights and biases completely? To fall back on our previous thought experiment consider that you are a neural network, and that you do completely update our weights and biases to each image. Then, when you see a new image of a cat *you completely overwrite any knowledge embedded in your weights and biases from other cat images you experience in the training set*. Instead, we only want a small shift of our weights and biases for each image. **This allows our network to learn generalizable patterns from each image, instead of just memorizing each image**. You may be asking "But how big of a shift should we make?". The size of the shift that we make is the **learning rate**, and there is no *best* learning rate. Instead, we have to experimentally find the best learning rate through hyperparameter tuning. 

## The Bias vs. Variance Tradeoff

Let's change gears for a second. We will take a moment to talk about how error is introduced, and the different types of error. 

Most errors in neural networks come for a poor balance between Bias and Variance. The trade-off between bias and variance is a fundamental concept in machine learning. It refers to the relationship between a model's ability to detect patterns that exist in training data (maintaining a low bias) without becoming so overly specific that it can't be applied to new data well (maintaining a low variance). We have hinted at this when we discussed overfitting, but if you are having issues with getting your model to improve, this is going to be well worth your time. 

But wait, what are bias and variance? Let's discuss. Imagine you play 3 games of darts. Suppose your games looks something like this:



image.png

It's pretty obvious the top left game is the best, but in what ways are the others bad? The bottom-left is just as clustered, but there seems to be a consistent offset from the bullseye. In other words, you were *biased* to throw tooo high, even though your darts have very low *variance* shot-per-shot. 

Meanwhile, in the top-right board, the darts are very spread out. Even though your darts were grouped generally around the middle, they were too varied to be any good. So, we say that this board has low bias, but high vairance. The bottom-right is clearly the worst -- it has high variance (a lot of spread) AND it is nowhere near the middle!

However, as the models variance increases, its ability to perform well on anything but the data on which it was trained decreases, so our validation loss increases greatly. 
<Details>

In general, after training for a while, we hope to see our models loss cuves look like the middle section. Both training and validation loss should be low, and, in general the validation loss will be slightly worse than the training loss. 

## The XOR Problem: Shallow doesn't cut it!

So, we have covered the difference between deep and shallow learning with perceptrons. But, we haven't discussed *why having hidden layers changes the network*. First, we are going to discuss at a high level the limitations of single-layer perceptrons, then we will walk through an example where a single-layer fails, but 2 layers succeeds. We know that neural networks are like logistic regression -- we do a bunch of linear regressions in parallel, and then each linear regression gets its own activation function (which is like the sigmoid function in logistic regression). The activation function lets us make a model that can do nonlinear classification. But, if we can do nonlinear  classification, then why do we need anything more? The simple answer is that sometimes (a lot of times), problems are too complex for just *one* curve, and going deeper allows our function to model more and more complex curves. But, what does it mean to have a more "compelx" function? And, what types of problems are too complex for one layer? In this next example, we are going to explore an example wherein one layer is too simple, and we need two layers. 

We are going to examine the XOR function and investigate why it cannot be solved by a single layer perceptron. **An important detail**: When we talk about a perceptron, we are talking about the layer *without* any activation function. Although we don't use one without the other in neural networks, the perceptron itself doesn't include them. 

XOR is a simple function that stands for "exclusive or". Let's start by defining what "or" means as a mathematical function. "OR" is a function that takes in two inputs, and has one output. The inputs and outputs are only allowed to be binary (0 or 1). When either or both of the inputs are 1, then the output is 1, else the output is 0. The following function implements "or". Take a second to go through it, to make sure you understand. 

```python
def OR(a,b):
  if a == 1:
    return 1
  elif b == 1:
    return 1
  elif a == 1 and b == 1:
    return 1
  else:
    return 0
```

Now, this is hardly the most efficient way to write OR, but it does make exatly what is happening explicitly clear. 

Let's test it to make sure it works.

```python
print(OR(1,0))
print(OR(1,1))
print(OR(0,1))
print(OR(0,0))
```

Exclisive or is very simmilar to OR, except that exlusive or is only 1 when *one of the inputs* is 1. If both inputs are 1, exclusive or returns 0. Let's write that function below. 

```python
def XOR(a,b):
  if a == 1 and b == 1:
    return 0
  elif b == 1:
    return 1
  elif a == 1:
    return 1
  else:
    return 0
```

Again, let's test it out.


```python
print(XOR(1,0))
print(XOR(1,1))
print(XOR(0,1))
print(XOR(0,0))
```

Good, our XOR function works -- its pretty simple. If a function is simple enough for us to implement it in a few lines of code, we can definetly model it with a neural network. So, let's take a deeper look at XOR.

No! To be linearly separable, you have to be able to draw a straight line between both classes (the 0's and the 1's). No single dotted line can separate the data here into two discrete classes.

Think about it this way: each perceptron unit (each single linear regression) can draw one dotted line. Now, we can have a single-layer of 2 perceptron units that draws the two lines we need to separate the data, but *we still need a second layer to combine them*. 

To understand this better, look at the below figure.

That's the power in deep learning. The first layers draw very rudimentary decision boundaries that aren't nearly complex enough to model most problems. But, adding a second layer allows the model to combine those low-level features into slightly more abstract features. As you go deeper and deeper, you can draw more and more complicated descision boundaries, and solve problems that you couldn't solve before. 

This process of finding simple features (drawing simple lines) in the first layer and combining them to make more complex features allows the neural network to automate the process of **feature extraction**. ## Let's Repeat the Logistic Regression Problem with a Multilayer Perceptron

Whew. After some feature engineering, we were finally able to get our logistic regression model to fit well to our data. However, as you saw, this isn't always a straightforward task. Imagine if we hadn't told you that age/performance was a good way to transform your dataset. That would have been really hard to realize on your own. Feature engineering is difficult and not straightforward. Frankly, it can be a pain to feature engineer a dataset so that a shallow algorithm can make sense of it. 

That's why deep neural networks are so important. They automate the process of feature engineering; they figure out how to combine the input data you give it. This means that our data **does not have to be linearly separable**. Let's use a feed-forward neural network (also called a multilayer perceptron) to solve the same problem as before. 

```python

from keras.models import Sequential
from keras.layers import Dense

x = data[['age','performance']]
y = data['y']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state = 101)

model = Sequential()
model.add(Dense(20, input_dim=2, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100, batch_size=10) #fit ANN

accuracy = model.evaluate(x_test, y_test)
print(round(accuracy[1]*100,2))
```

```python
#make predictions on the decision boundry points
predictions = model.predict(data_db[['age','performance']])
y_db = np.around(predictions)
data_db['y'] = y_db

fig, ax = plt.subplots( nrows=1, ncols=1,figsize=(15,10))

#Plot decision boundry
plt.scatter('age','performance',c='#ffbdbd',s=1,data=data_db[data_db.y == 1])
plt.scatter('age','performance',c='#b0c4ff',s=1,data=data_db[data_db.y == 0])

#Plot employee data points 
plt.scatter('age','performance',c='#ff2121',s=50,edgecolors='#000000',data=data[data.y == 1])
plt.scatter('age','performance',c='#2176ff',s=50,edgecolors='#000000',data=data[data.y == 0])
plt.ylabel("Performance Score",size=20)
plt.xlabel('Age',size=20)
plt.yticks(size=12)
plt.xticks(size=12)
```

# Limitations of Deep Learning? When to opt for Shallow Learning instead.

Deep learning is a very popular and powerful tool for solving a wide variety of problems. Additionally, if a problem can be solved with shallow learning, deep learning can also solve it. However, just because you *can* use deep learning on almost any problem, this doesn't always mean you *should*. 

Deep learning does have some limitations that should be considered when deciding whether to use it or not. Here are 4 limitations of deep learning that are good to keep in mind when you are deciding what algorithm to apply to a given problem:


## When your training data is limited.

When you have limited data, it may be more appropriate to use a shallow learning approach instead of deep learning. The main reason for this is that deep learning models require large amounts of data to train effectively. When you only have a little bit of data, a deep learning model may fail to learn well because it may *overfit* the training data. We have already discussed overfitting in relation to variance and the loss curve, so you should already have a good idea of how this works. Remember our analogy for how machine learning is similar to how we learn in school? Imagine if your teacher only gives you a couple examples for a complex topic in math. You spend a ton of time studying everything about those two examples and you know them in and out. Then your teacher gives you another problem -- same concept, but it's not quite the same as your examples. What happens? It's pretty likely you will mess up the problem, because you just memorized those two examples -- you didn't have enough data to really learn the process. 

On the other hand, shallow learning programs are less complex and require less data to train. Shallow learning may be more suitable for small datasets because they are less prone to overfitting and may still provide good performance.

## When your compute power is limited.

Something you may have noticed while running through the examples in this module is the drastic difference in time that it took to train the logistic regression vs. how long it took to train the neural network. Neural Networks, compared to shallow learning algorithms, take a really long time to train. Depending on how good of a computer you have, sometimes it simply isn't feasible to train a neural network. 

In these cases, it's often preferable to use shallow learning approaches. 


## When you want easier deployment.

As you will experience firsthand shortly, neural networks are often difficult to train. They have many different hyperparameters that you can tune, and finding good values for all of these can be a hefty task. If the problem is simple enough to solve with shallow learning, it's often quicker and easier to just use it instead. 

## When Interpretability is important.

Deep learning models can be difficult to interpret, making it hard to understand why a particular decision was made. This can be a problem in applications where transparency and accountability are important.

However, shallow learning is often easier to understand as it consests of more simple functions. Some algorithms can even show you the exact math (in a human readable format) that it used to produce its output.

When interpretability is important, it may be more appropriate to use a shallow learning approach instead of deep learning. The main reason for this is that deep learning models can be difficult to interpret, making it hard to understand why a particular decision was made. This can be a problem in applications where transparency and accountability are important

## Concluding Remarks

The big takeaway here is that you shouldn't treat deep learning as a silver bullet. Although it probably *can* solve whatever problem you are trying to solve, sometimes it may be overkill. In these cases, it may be more appropriate to use a shallow learning approach instead of deep learning. Shallow learning algorithms, such as logistic regression or decision trees, are simpler and easier to interpret than deep learning models, and they can often achieve comporable or better performance on smaller datasets. Shallow learning algorithms are also computationally less expensive than deep learning models, making them easier to use and, often, more suitable for applications where real-time processing is required.
