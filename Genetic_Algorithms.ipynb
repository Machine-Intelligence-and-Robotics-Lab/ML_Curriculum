#Biological Inspriation for the Genetic Algorithm

To begin, it's time for a quick refresher from biology class. In this module, we are going to discuss genetic algorithms. Genetic algorithms are a class of a larger set of algorithms (called evolutionary algorithms) that take inspiration from biology. In genetic algorithms, solutions to a problem are represented as chromosomes, just like in real genetics. These chromosomes are then subjected to a process of selection and recombination, mimicking the process of natural selection. By repeating this process over many generations, the algorithm can converge on an optimal solution to the problem at hand. Since the motivation behind genetic algorithms derive from genetics and evolution, it's probably best we start there.

## What are Genetics?

Genetics is the study of how traits are inherited and expressed in living organisms. Each organism has a unique set of instructions, encoded in its DNA, that dictate its physical and behavioral characteristics. These instructions are passed down from one generation to the next, through the process of reproduction.

Gregor Mendel  often referred to as the father of modern genetics. He discovered the basic principles of genetics through his work with pea plants. He found that certain traits, such as flower color or plant height, were inherited in predictable patterns from one generation to the next. This means that if you know the traits of the parents, you can make predictions about the traits of their offspring.

One way to visualize the inheritance of traits is to use Punnett squares. A Punnett square is a simple diagram that allows us to make predictions about the children's genetics based on the parent's genetics. It's a chart that organizes all the possible combinations of genetic information that can come from each parent.

To make a Punnett square, we write the possible alleles (versions of a gene) for each parent along the top and side of a square grid. Then we fill in the boxes with the possible combinations of alleles that their offspring could inherit. For example, if we are looking at flower color in pea plants, and the parents have the alleles for purple (P) and white (p), we would write P and p across the top and side of the Punnett square, and then fill in the boxes with PP, Pp, pP, and pp, representing the possible combinations of alleles that their offspring could inherit. Over time, genetic variations compound within a population through the process of mutation. Mutations are random changes in DNA that can lead to new traits, some of which may be beneficial, harmful, or have no effect at all. ## What is Evolution?

Evolution is the idea that species come about through the process of pre-existing species adapting to their environments. The mention of adaptation refers to natural selection, known colloquially as "survival of the fittest". Natural selection is the process by which certain traits become more or less common in a population over time, depending on their impact on an organism's survival and reproduction. Organisms with traits that are better suited to their environment are more likely to survive and reproduce, passing on their advantageous traits to their offspring. Over time, this can lead to the evolution of new species that are better adapted to their environment.

The evolution of the Peppered Moth species provides a clear illustration of natural selection at work. During the Industrial Revolution, the widespread soot pollution dirtied up the white trees and buildings that the Peppered Moths used for camouflage, causing them to stand out to birds and leading to a decline in their population numbers. However, some of the moths had a slightly darker hue than others, making them less visible to predators. As a result, over a few generations, the population became dominated by the darker variant, which was better adapted to the environment, and the moth population's numbers began to level out and increase once again. This example highlights the critical role of natural selection in determining the success of a population based on their ability to adapt to changes in their environment.

# Evolutionary Algorithms

Evolutionary algorithms are a type of algorithm that use the principles of natural selection and genetics to solve complex problems. In these algorithms, potential solutions to a problem are represented as individuals in a population, just like in real genetics. The algorithm then applies selection, crossover, and mutation to these individuals to simulate the process of natural selection. Through many generations of this process, the algorithm can find a solution to the problem at hand.

The analogy between evolutionary algorithms and natural selection is clear. In both cases, the goal is to produce the "fittest" individuals through a process of selection and reproduction. However, evolutionary algorithms put us in the driver's seat. We can manipulate the parameters of the selection process, have the ability to simulate many generations of evolution in a short period of time, and can apply these principles to a wide range of problems outside of biology.

In the case of genetic algorithms, we will use 5 steps to simulate the process of evolution. They are:

1. Initialization: establishes a gene pool for the algorithm to pull from. A set of members are created with a wide variety of values depending on the context.

2. Selection: members are sent through a function, called the fitness function, which tests their viability for the solution. The fitness function depends strongly on the situation and expectations. The best members are sent to the next steps. This function is analogus to the process of natural selection.

3. Crossover: new members are generated with the values of the best members from the selection step. These new members make up a full population for the next cycle. This is analogus to the process of genetic recombination via sexual reproduction we see in biology. Instead of following the laws of Mendelian genetics, however, artifical genes recombine according to rules we define.

4. Mutation: A random set of chromosomes withing individual members are changed. We do this in order to keep the gene pool wide. Without this step, our population doesn't have much room to evolve in a way that is vastly different from the parents. Essentially, without mutation, our population becomes inbred: it begins to lack the genetic diversity needed to develop radically diffrent genetics than those of the parent's generation, which damages the algorithms ability to explore new approaches to a problem.

5. Termination: We stop creating new generations once some number of generations that we set have been created, or we consider the solutions satisfactory.

## How are Genetic Algorithms similar to biological evolution and genes?

At this point, we have explored well the analogy between biological genetics and genetic algorithms. A gene is some list of numbers that represents a solution to some problem we want to solve. Each number in a gene is referred to as a chromosome. Genetic algorithms change a starting population of genes by crossing the genes over with other members of the population, and then mutating some of the chromosomes. Then, we apply a fitness function to determine which genes are most fit to solve the problem. We use those to form the next generation, and we repeat. Eventually, our genes become fit solutions to whatever problem we were trying to solve.

It's worth noting that genetic algorithms are not a perfect analog for genetics and evolution. They operate on a much shorter timescale than biological evolution, and don't take into account some of the more complex factors that influence genetic inheritance. However, genetic algorithms still take heavy inspiration from biological genetics, and it's useful for us to study genetic algorithms through the lens of biology.

## The Steps of a Genetic Algorithm

Genetic algorithms function by evolving a gene that represents a workable solution to a problem we define. Just as natural selection guides evolution to create creatures fit to live, The functions we define guide our algorithm to create solutions fit to solve our problem. Genetic algorithms can be defined in a series of steps:

1.   Initialization - We initialize a starting populations with random chromosomes.

2.   Fitness - We determine the fitness of the population, which is like the ability of each organism to survive in its environment.

3.   Selection - We select the most fit chromosomes from the population to evolve toward a more fit generation of children.

4.   Crossover - We generate offspring based off the parents that were selected in the previous step. This is like reproduction, but instead of genetic crossover represented by a punnent square, we get to define our own crossover method.

5.   Mutation - In order to promote diversity in the population, we randomly mutate some of the chromosomes in each gene.

6.   Check Termination Conditions - Unlike evolution, which is continuous, we have to choose a stopping point. We generally choose to stop when:
      1.   we see enogh improvement in the solution,

      2.   We reach a pre-defined number of generations we want to create with our algorithm

      3.   Our fitness function reaches the minimum value

### Initialization

The first step in creating a genetic algorithm is to initialize the population we will be using. A population is only limited by how much power can be applied towards the processing: it could be 5 members, or it could be 500. It all depends on what our computer can handle. Each member of the population is effectively an array of numerical values that can be evaluated by the fitness function. The fitness function will be determined by our goal -- whatever problem we are trying to solve. That goal is established based on the situation, and there is a large variety of even simple goals that we can engineer the genetic algorithm to solve towards.

There are two common methods for initializing a population: random initialization and heuristic initialization. With random initialization, we create a population using completely random individuals or values. For example, if we let our genes be size = 3, we may generate 10 members with random values between x1 and x2 for each value in their array. The range of values for x1 and x2 are determined by the problem itself. This approach provides the opportunity for the most diversity, which can help ensure that the population has a chance to evolve in a lot of different ways. For an example, let's say we want to reach a point in 3D space dictated by x, y, and z values. X is the left and rights axis, Y is the depth axis, and Z is the height axis. Suppose we want our algorithm to get as close as it can to (11, 13, 4) within a 20x20x20 cube. We can generate 10 members with random values between 0 and 20 for each value in their array. With this, they should be fairly spread around within the box, and our algorithm can find its way to (11, 13, 4) using it's fitness function to judge distance.

On the other hand, with heuristic initialization, we create a population using a key or rule that lacks complexity. In other words, if we already have a good idea of how the solution should look, or have a solution that works partly, but could be better, we can use this solution(s) as a starting place for our genetic algorithm. However, this can lead to little diversity in the population, which may prevent the algorithm from performing as well as it could. Additionally, if we don't know a good starting place, we can't use heuristic initialization.

In pracice, it's best to use a combination of random and heuristic initialization. We start with heuristic, filling the initial population with those solutions that we already believe to be good, and add in random genes after that.

### Fitness

The fitness function is a crucial aspect of the genetic algorithm, as it generates a fitness score that reflects an individual's "likelihood of survival". This score is then used to rank the members of the population and select the best candidates for reproduction. To determine the fitness score, each member is tested using an equation to determine how close they are to the goal. In our example of reaching a point in 3D space, the fitness function would find the distance between the location described by our new gene and the point(11, 13, 4). We choose the top performers, those with the smallest resulting overall fitness score, to move on to the next cycle.

For instance, if we generated 10 members with random values between 0 and 20 for each value in their array, and the goal was to reach the point (11, 13, 4) within a 20x20x20 cube, we could take the top two members that were closest to the goal. Let's say that the members with the values (8, 15, 7) and (4, 6, 2) had fitness scores of 8 and 16, respectively. Although they weren't very close to the goal, they were the top performers among the population, and thus would be selected for reproduction.

### Selection

After the fitness function has been applied to each member of the population, the next step is to select individuals for reproduction. The idea is to choose the fittest individuals as parents in order to produce even fitter offspring. However, it is also important to maintain genetic diversity in the population to avoid converging to a suboptimal solution. One method of achieving this balance is through Roulette Wheel Selection.

In this method, each individual is assigned a slice of a roulette wheel, with the size of the slice proportional to its fitness score. The wheel is then spun twice, and the individuals whose slices the wheel stops on are selected as parents. This process can be repeated to select as many parents as needed.




image.png

The selection process is based on probability, where individuals with higher fitness scores are more likely to be selected, but there is still a chance for lower fitness individuals to be chosen. This helps maintain diversity in the population, as it allows for the possibility of "hidden gems" to be discovered among the lower fitness individuals. In other words, it provides a good balance between exploitation (selecting the fittest individuals) and exploration (maintaining genetic diversity), a key concept motivating the genetic alogorithm approach.

### Crossover

Now, we  cross the genes in order to move toward a good solution. There are a few ways to do this. You can just cut the values in half, or split them a couple ways. The choice can depend on the number of different values the members contain, how you prefer they learn, and any constraints imposed by the problem itself.

#### Single and Double Crossover

A simple crossover method is to just cut the values in half (or at some other location) and stitch them back together on other members. This is called single-point crossover. Double-point crossover is similar, we just cut them in a few places to stitch back. The choice between the two can depend on the length of the arrays.

A single crossover on a set of 6 arrays might cut them in half to cross them as shown below;

A double crossover will slice them twice. A double crossover might cut the first two and the last two from an array of 6 values like so;

DoubleCrossover.png


If you want, you can keep doing this 3, or 4, or however many times you wish. This is called multi-point crossover.

In our example, our two winners were (8, 15, 7) and (4, 6, 2). For a single crossover, let's just cut the first value off and replace it with the other, so we have (4, 15, 7) and (8, 6, 2) now. A double crossover would add another cut, and the only other place we could cut is the last value so we would get (4, 15, 2) and (8, 6, 7).

#### Uniform Crossover

A uniform crossover is a little more complicated, but works better for some applications. Uniform crossover uses some process to select genes to cross, either random, or depending on the success of the individual values.

### Mutation

Mutation is a way to introduce a bit of chaos into our solution. More formally,it allows us to explore a larger range of solutions. There are many of ways to implement mutation, and which way we implement it depends on our specific problem. One method of mutation is Bit Flip Mutation, where random bits are chosen and their values are inverted. Another approach is Random Resetting, which involves replacing a chromosome with a random value. Another method is Swap Mutation, where random bits are chosen, and their positions are swapped. In Scramble Mutation, a section of bits is randomly chosen, and their values are scrambled and repositioned. Lastly, in Inversion Mutation, a section of bits is selected, and their order is inverted.

The goal of mutation is to introduce some randomness and diversity to the population to fill out missing gaps and prevent stagnation. The number of bits that are changed can be influenced by a mutation rate value, which determines the likelihood of a mutation occurring in any given iteration. By introducing a bit of randomness into the genes, we can explore new and possibly more optimal solutions to the problem at hand.

### Termination

The last step in implementing a genetic algorithm is to define a termination condition. This can be based on the number of rounds or the amount of time the process has taken, or how close the members have gotten to the goal. The choice of termination method depends on the complexity of the goal and the time available. For more complicated or open-ended goals, a time or round-based termination is more appropriate, while simpler goals can be evaluated based on how close the members come to the desired threshold. In either case, the ultimate goal is to select the best solution found so far and stop the program. This can be determined either by runtime or performance, with the program stopping when it has run for too long or when the best solution has been found and designated as the final solution.

# When to use a Genetic Algorithm

Genetic Algorithms are a powerful class of optimization algorithm, capable of finding workable solutions to problems for which we cannot otherwise find solutions. However, they are not ideal in all situations; sometimes other approaches are better suited to task. In this section, we will cover three general conditions you want to have when deciding whether to use a genetic algorithm.

## When the problem is discrete

Another word for discrete is countable. So, when we talk about problems that are discrete, we really mean problems whose input variables for which we are optimizing values need to each only have a finite set of options they could be. For example, you can apply a genetic algorithm to choose which cities that Amazon should direct its shipping vans through (there are a finite number of cities), but it can't as well be applied to determine what speed the van should go (because speed is continuous -- you can go 60 mph, or 60.1 mph, or 60.01 mph, etc. There are infinite number of values for speed).

## When you are looking for a solution to a specific problem instead of pattern recognition

Last module, we learned that neural networks extract features from data and learn the patterns expressed in those features. This is a very powerful and prevalent form of machine learning. Genetic Algorihms don't work like this. Instea of using data to learn about the solution space, Genetic Algorithms start with a randomly initialized population. We can think of the population as points on our solution curve. Each iteration, the algorithm selects those points that represent the best solutions and performs crossover and mutation. Crosover and mutation spawn more solutions (more dots) on our surface. The algorithm repeats until it finds a solution that we think is good enough.

## When you aren't looking for *the* optimal solution, just a solution that works

Notice that I said "good enough", not the best. We already covered that optimizing algorithms have no way to know if they have found the absolute minimum, or just a local minimum. On top of that, genetic algorithms don't search for a solution by iterating directly toward a minimum. Rather, they use mutatuion and crossover to randomly spawn children near the parents. Because of this randomness, even when they have the same initial settings, each time you run a genetic algorithm, you can end with a different solution. Therefore Genetic Algorithms are good when you aren't too concerned with finding the very bottom of the optimization valley, but instead will be satisfied with a solution that is good enough.

## What's the traveling salesman problem?

Meet Ana. Ana lives in the suburbs of Atlanta with her family. Atlanta is a good place for Ana and her family to live, given her line of work. She is a traveling salesman, and the Atlanta airport is one of the busiest in the world, which makes Ana's travel time from the airport to his home quite swift. However, the time in the air is a much different story.

Family time is important to Ana, and she doesn't like to spend longer traveling than she absolutely has to. Each month, her company gives him a list of cities that he is supposed to travel to. She is allowed to pick the order in which he travels. Let's assume that Ana can always travel from any city she is in to any other city without waiting for layovers/waiting on a plane. Also, Ana cannot visit any city more than once (her company won't pay for her travel otherwise). How can she figure out the order in which he should travel to the various cities?

If she only has to visit a small number of cities, it's pretty easy to figure out the quickest route. For example, if she has to visit three cities (say Chicago and Denver),  there are only two possible routes, and they are both the same length. She can either leave Atlanta to Chicago, then to Denver, then home, or she can go to Denver then Chicago.

Now, add a 4th city. How many potential routes could Ana take? The answer is 6. For 5 cities, the answer is 120. At this point, you probably see where this is going. For 10 cities, there are over 3 million different routes that Ana could take. Obviously, it is not feasible for Ana to sit down and figure out which path is the shortest distance for *3 million* different options. So, what should she do?

We could use a computer to check all 3 million routes one-by-one. This is called a *brute force* approach. But, this doesn't scale. Say she gets a busy month and her company sends her to 21 cities. That means there are over *51 quadrillion* routes to check. That's more routes than there are grains of sand on the earth. Even the worlds fastest computer would need multiple days, if not years, to check every route. Obviously, brute forcing this problem isn't going to get us very far.
```python
import numpy as np, random, operator, pandas as pd, matplotlib.pyplot as plt

class City:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, city):
        xDis = abs(self.x - city.x)
        yDis = abs(self.y - city.y)
        distance = np.sqrt((xDis ** 2) + (yDis ** 2))
        return distance

    def __repr__(self):
        return "(" + str(self.x) + "," + str(self.y)

class Fitness:
    def __init__(self, route):
        self.route = route
        self.distance = 0
        self.fitness= 0.0

    def routeDistance(self):
        if self.distance ==0:
            pathDistance = 0
            for i in range(0, len(self.route)):
                fromCity = self.route[i]
                toCity = None
                if i + 1 < len(self.route):
                    toCity = self.route[i + 1]
                else:
                    toCity = self.route[0]
                pathDistance += fromCity.distance(toCity)
            self.distance = pathDistance
        return self.distance

    def routeFitness(self):
        if self.fitness == 0:
            self.fitness = 1 / float(self.routeDistance())
        return self.fitness


def createRoute(cityList):
    route = random.sample(cityList, len(cityList))
    return route


def initialPopulation(popSize, cityList):
    population = []

    for i in range(0, popSize):
        population.append(createRoute(cityList))
    return population


def rankRoutes(population):
    fitnessResults = {}
    for i in range(0,len(population)):
        fitnessResults[i] = Fitness(population[i]).routeFitness()
    return sorted(fitnessResults.items(), key = operator.itemgetter(1), reverse = True)


def selection(popRanked, eliteSize):
    selectionResults = []
    df = pd.DataFrame(np.array(popRanked), columns=["Index","Fitness"])
    df['cum_sum'] = df.Fitness.cumsum()
    df['cum_perc'] = 100*df.cum_sum/df.Fitness.sum()

    for i in range(0, eliteSize):
        selectionResults.append(popRanked[i][0])
    for i in range(0, len(popRanked) - eliteSize):
        pick = 100*random.random()
        for i in range(0, len(popRanked)):
            if pick <= df.iat[i,3]:
                selectionResults.append(popRanked[i][0])
                break
    return selectionResults



def matingPool(population, selectionResults):
    matingpool = []
    for i in range(0, len(selectionResults)):
        index = selectionResults[i]
        matingpool.append(population[index])
    return matingpool


def breed(parent1, parent2):


    geneA = int(random.random() * len(parent1))# get a random point in parent 1
    geneB = int(random.random() * len(parent1))# get another random point in parent 1

    startGene = min(geneA, geneB)#the starting Gene for our swap will be the smaller random point
    endGene = max(geneA, geneB)#the ending Gene for our swap will be the larger random point

    #initializing the size of our child
    child = [None]*len(parent1)

    for i in range(len(parent1)): #looping through all the genes
      if (i >= startGene) and (i <= endGene):#if we are within the range that comes from parent 1
        child[i] = parent1[i]#add that gene to our child

    childP2 = [item for item in parent2 if item not in child]#get genes from parent 2 by searching for one's that the child doesn't already have

    for i in range(len(parent2)): #looping through all the genes
      if not((i >= startGene) and (i <= endGene)):#if we are not within the range that comes from parent 1
        child[i] = childP2.pop(0)#put the next chromosome from the 2nd parent to the child


    return child


def breedPopulation(matingpool, eliteSize):
    children = []
    length = len(matingpool) - eliteSize
    pool = random.sample(matingpool, len(matingpool))

    for i in range(0,eliteSize):
        children.append(matingpool[i])

    for i in range(0, length):
        child = breed(pool[i], pool[len(matingpool)-i-1])
        children.append(child)
    return children



def mutate(individual, mutationRate):
    for swapped in range(len(individual)): #for each element in the individual
        if(random.random() < mutationRate):#randomly check if this gene will be mutated
        #To deal with the fact that every city can only be used once, we mutate by swapping the location of
        #two chromosomes on the same individual.
            swapWith = int(random.random() * len(individual)) #find another chromosome in the same individual gene

            #perform the swap
            city1 = individual[swapped]
            city2 = individual[swapWith]

            individual[swapped] = city2
            individual[swapWith] = city1

    return individual


def mutatePopulation(population, mutationRate):
    mutatedPop = []

    for ind in range(0, len(population)):
        mutatedInd = mutate(population[ind], mutationRate)
        mutatedPop.append(mutatedInd)
    return mutatedPop



def nextGeneration(currentGen, eliteSize, mutationRate):
    popRanked = rankRoutes(currentGen)
    selectionResults = selection(popRanked, eliteSize)
    matingpool = matingPool(currentGen, selectionResults)
    children = breedPopulation(matingpool, eliteSize)
    nextGeneration = mutatePopulation(children, mutationRate)
    return nextGeneration



def geneticAlgorithm(population, popSize, eliteSize, mutationRate, generations):
    pop = initialPopulation(popSize, population)
    print("Initial distance: " + str(1 / rankRoutes(pop)[0][1]))

    for i in range(0, generations):
        pop = nextGeneration(pop, eliteSize, mutationRate)

    print("Final distance: " + str(1 / rankRoutes(pop)[0][1]))
    bestRouteIndex = rankRoutes(pop)[0][0]
    bestRoute = pop[bestRouteIndex]
    return bestRoute



cityList = []

for i in range(0,25):
    cityList.append(City(x=int(random.random() * 200), y=int(random.random() * 200)))

geneticAlgorithm(population=cityList, popSize=100, eliteSize=20, mutationRate=0.01, generations=500)



def geneticAlgorithmPlot(population, popSize, eliteSize, mutationRate, generations):
    pop = initialPopulation(popSize, population)
    progress = []
    progress.append(1 / rankRoutes(pop)[0][1])

    for i in range(0, generations):
        pop = nextGeneration(pop, eliteSize, mutationRate)
        progress.append(1 / rankRoutes(pop)[0][1])

    plt.plot(progress)
    plt.ylabel('Distance')
    plt.xlabel('Generation')
    plt.show()


geneticAlgorithmPlot(population=cityList, popSize=100, eliteSize=20, mutationRate=0.01, generations=500)
```
