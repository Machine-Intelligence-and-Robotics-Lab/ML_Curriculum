#What is Natural Language Processing?

Natural Language Processing (NLP) is the subset of machine learning that deals with teaching computers how to understand, interpret, and/or use human language. At its core, NLP is focused on the interaction between computers and people through natural language. When we say natural language, we are referring to the the style of language that we normally use to communicate. When you talk to your mom, your doctor, your friend, or your teacher, you are using natural language. Natural language itself is incredibly complex and nuanced, with countless variations and subtleties that can make it difficult for computers to accurately interpret. With all the different slang words, words with multiple meaning, and context-dependent word choice, creating an algorithm to make sense of the way we speak has been one of the most challenging tasks machine learning engineers have had to face. Even so, NLP has made tremendous strides. In recent years, we have seen the advent of now-famous virtual assistants like Siri and Alexa, chatbots like ChatGPT, and web search engines like Google and Bing, all of which are NLP algorithms. NLP has begun to revolutionize the way we interact with technology by allowing humans to communicate with computers in a way that feels relatively natural, and it shows no signs of slowing down.

#First, A Histroy Lesson!

Before jumping into the nitty-gritty of NLP, let's take a second to disuss the history of NLP -- how we got to where we are today. We will give a high level overview of how NLP was concieved as a field, the series of developments that followed, and what the state-of-the-art in NLP looks like.

##Early attempts at NLP

As we have mentioned, teaching computers to understand human language is no easy task. Language is full of nuances, context, and cultural references that can be challenging for computers to interpret and process. Understanding language requires a deep understanding of context and the ability to draw conclusions from implicit and explicit cues. Sometimes even other humans fail to pick up on the subtlties of language. Despite this,  over the last century, researchers have made significant progress in developing NLP models that can perform a wide range of tasks using advanced machine learning algorithms and techniques.

NLP has a long history, dating back to the 1940s when researchers first explored how computers could understand and generate human language. The first motivation for NLP arose from World War II -- people began to realize the importance of being able to automatically translate between languages. Initially, rule-based systems were used to develop grammars and dictionaries that defined the structure and meaning of language. They tried to hand-code an algorithmic representation of grammar and word-meaning. It didn't take long for researchers to realize that this approach wasn't going to cut it -- manually developing rules to model the entire english lexicon was quite a bit of work, to put it mildly. So, in the 1950s and 60s, scientists developed statistical NLP methods, whcih lead to the first machine translation systems -- systems that could automatically translate text between languages. However, even these statistical models struggled to grasp the sheer complexity of language. New techniques were still needed.

Both symbolic and statistical approaches to NLP continued for some time, but in the 1980s and after, statistical models began to pull ahead. Words were often pre-processed into their root words through processes called stemming and lemmatization. Early techniques such as  frequency matching (grouping documents by how frequently certain words appear) became prominent.

## The First NLP Neural Networks

In the early 2000s, a major development in NLP was the introduction of recurrent neural networks (RNNs) to predict the probability of a word given its preceding words. These models allowed for more accurate and robust language modeling, and were used in tasks such as speech recognition, machine translation, and text classification. However, they were still far from perfect. They were fairly unstable, and very difficult to train. Additionally, they struggled to understand long-range dependencies and context in sentences. In other words, if some input was talking about the same topic for multiple sentences, RNNs started to forget what the topic of conversation was.  

A few years after that, a variant on the RNN, called the Long-Short-Term Memory (LSTM) network, was developed to overcome the RNN's struggle with long-range dependencies and provided a partial solution to some of the problems that RNN's introduced. They performed better than typical RNN architectures, but they still had problems with complexity and were very prone to overfitting.

Around the same time, the word2vec class of NLP models popularized the word embedding (a technique where natural language is represented as a multi-dimensional vector). We will discuss word2vec in much more detail later.

In the mid-2000s, sequence-to-sequence models were developed, which are neural networks that can map fixed-length input sequences to fixed-length output sequences. This allowed NLP architectures to adapt better to natural language, since natural sentences are sequential (the order of words in a sentence matters).

Later in the 2000s, attention mechanisms were introduced, which allow neural networks to selectively focus on different parts of an input sequence when generating an output sequence. This led to significant improvements in tasks such as machine translation and text summarization.

A few years later, attention mechanisms were introduced and gave rise to the transformer model -- an architecture now used in most modern NLP models. The last 5-6 years has seen an explosion of NLP technology based on the transformer architecture; models like BERT which run Google's search engine and GPT-3, the model behind ChatGPT are transformer architectures. Transformers are so popular, in part, because they mastered **transfer learning** and **multi task learning**.  


## Multi Task Learning Neural Networks

Multi-task learning is a type of machine learning technique that involves training a neural network to perform multiple related tasks simultaneously, rather than training separate networks for each task. This approach can be more efficient and effective than training individual networks for each task, since the network can learn to share information and resources between the different tasks. A common approach to implementing multi-task learning is *transfer learning*. Transfer learning is a machine learning scheme where the model is trained on is different from the one it will be used for. The model trains on a task that is relatively easy to implement in hopes that it will extract useful patterns for many different, often more useful, tasks later on, with only a little fine-tuning.

That sounds kind of complicated, I'll admit, but the concept is probably more farmiliar than you realize. Have you ever seen the old karate kid movie? If you haven't, that's okay, I'll quickly summarize the important points for you. The movie is about a high school student, Daniel, who wants to learn karate. He comes under the membership of an elderly karate master who lives at his apartment complex, Mr. Miyagi, who takes an odd approach to teaching him. Mr. Miyagi begins "training" Daniel by having him do chores -- waxing his car collection, painting the fence, and sanding the floors. Mr. Miyagi makes daniel perform the chores in a very repetitive fasion -- he is supposed to "wax-on" the cars with one small circular motion, then "wax-off" with the inverse, small swift circle. Daniel gets frustrated -- he doesn't see what these repetitive chores have to do with karate. Mr. Miyagi then dramatically reveals his training strategy -- he strikes at Daniel, telling him to "wax-on" to parry the strike. To Daniel's surprise, his muscle memory kicks in, and he parrys the strike perfectly using the same technique he used in waxing the car. In other words, the muscle memory that Daniel developed doing chores *transfers* to skill in karate. This, in essence, is what transfer learning is about. We give the neural network a simple task that allows it to learn the fundamentals about the data on which it is trained, just like how Mr. Miyagi gives Daniel the simple task of waxing  a car to allow him to learn the fundamentals of karate. In the movie, after Daniel has mastered the fundamentals through his chores, they begin practicing karate directly. Similarly, after we train our neural network on the training task, we train it on a small amount of data using our final task to fine-tune it to do exactly what we want it to.

In the context of natural language processing (NLP), multi-task learning can be used to train a single neural network to perform multiple related tasks, such as language modeling, named entity recognition, sentiment analysis, and machine translation. For example, the same baseline multi-task NLP model, with a little fine-tuning on each task, could predict the next word in a sentence, identify named entities (such as people or organizations), and determine the sentiment of the sentence (positive or negative).

# Word Embeddings: Graphing out words

So, we know what NLP is, and why we do it -- NLP is the field of ML that concerns itself with processing human language, and we use it for things like text summarization, sentiment analysis, chatbots, automatic translation, and much more. We have talked about some early attempts at NLP -- how NLP took two early approaches, symbolic and stochastic. Symbolic approaches mostly fell out of practice, and stochastic models became more prominent. Eventually, people began applying neural networks to NLP problems. In 2003, the field of NLP was revolutionized by a new approach to NLP: word embeddings. Unlike previous attempts at NLP, word embeddings borrow from structuralist linguistic theory, which posits that the meaning of words comes from the context in which they are used. According to this understanding of language, words don't inherintly mean anything -- rather, language is a structural system built with words.

Word Embeddings operate on this idea: instead of paying much attention to the words themseves, we train word embeddings by exposing it to the context in which words are used, and letting it learn the relationship between them.

But how do we encode this information? After all, the relationship between words is not just *one* thing -- it's the relationship between two or more things. The intuition behind how word embeddings solve this problem is actually quite simple: just graph the words out. Let's have an example.

Imagine you are my machine learner creating word embeddings. I am going to give you some words, and based on the context in which you typically see those words, you are going to place them on a graph.

First, I am going to give you the word "dog". Now, we only have one word, so you can put it anywhere, randomly. For our example, we are only going to use a 2D graph, but actual models that use *much* larger spaces (think a few hundred dimensions).

image.png

Now, I'm going to give you another word: "good". "Good" and "dog" are seen in the same sentences together all the time -- how often do we hear someone say "good dog"?. So, we should put those two words close together, since they are often used together:

image.png

This is the key idea behind a word embedding: words that are used together, are graphed together. If we subscribe to a structural linguist's understanding of language, then this distance between words -- how frequently they are used together, encodes the meanings of the words themselves. Given more words, and more copora in which to observe what words are used together, our model can build complex models of the relationships between words.

Let's add another word: "river". River is pretty different from the previous two words. Maybe they are occasionally used together, but not nearly as much as "good" and "dog" are. So, we should probably put that in a different part of the graph:

image.png

Similarly, we could probably graph "winding", and "water" fairly close to "river":

image.png

If I add the word "money"., it should probably go pretty far away from both groupings.

image.png

Now comes a problem. What if I ask you to graph the word "bank". Well, bank has two meanings: we could mean bank as in "river bank", or we could mean financial bank, as in "I put my money in the bank". "Bank" somehow needs to be close to "money" and "river", but that doesn't seem possible. Some words truly have multiple meanings, and our current approach to NLP can't account for this. Differnetiating between which meaning of words like "bank" is being used is called *Word Sense Disambiguation*, and it was a **major** roadblock for scientists in NLP for years. It wasn't until around 2017 that a powerful solution (called the attention mechanism) was invented that solved this problem in NLP. We will cover *how* attention does this later. For now, we will explore one of the most popular word emdedding architectures: word2vec. While you explore what word2vec can do, keep the problem of word sense disambiguation in the back of your mind, and try to think of potential solutions *you* might have tried to deal with this problem, had you been an NLP researcher in the mid 2010's.

To begin, we are going to load up a word2vec model trained on the Google News dataset.

```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

Let's have a look at one of the vector word embeddings in the word2vec model. Let's look at "king".

```python
vec_king = wv['king']
print(vec_king)
```

What are the dimensions of vec_king?

```python
len(vec_king)
```

Whereas our earlier example, where we graphed the words "dog", "good", "river" and such used 2 dimensions, this model uses 300. This is where the "300" in "word2vec-google-news-300" comes from -- it corresponds to how many dimensions the word2vec model has to represent words.

One of the fun things about word2vec is that, since it represents words as a vector, we can do math with them! A classic example of the power of word2vec's encoding is the vector KING - MAN + WOMAN = QUEEN. With word2vec, we should be able to perform these opertions, and see that this relationship holds true. Let's perform the operation, and then print out the 10 words in word2vec's dictionary that are most similar to our new vector.  

```python
new_vector = wv['king'] - wv['man'] + wv['woman']
print(wv.most_similar(positive = new_vector))
```

It seems that the most similar word is king... but the second most similar is queen! This is the same finding as the group who first illustrated the KING - WOMAN + MAN in word2vec. However, it seems a bit misleading -- if the most similar word is just the one we started at, then did we do anything? How do we know there is actually any meaningful relationship encoded in our vectors at all? Well, let's look at the king vector and see what the difference is.

```python
wv.most_similar(positive = wv['king'])
```

Notice that the number associated with each word represents how similar each word is to the vector in question. The closer they are to 1, the closer the words are in our vector space. When we look at the represenation of king, we find that "king" was exactly equal to "king (score of 1). "Queen" was .65 similar to king.

When we did our math and calculated king - man + woman, we found that "king" got less similar (.844) and queen got more similar (.73). So, we know that, even though queen isn't more similar than king, we moved *in the direction towards queen*.

When we double the "man" and "queen" vectors, we get that the queen vector is the most similar.

```python
new_vector2 = wv['king'] - 2*wv['man'] + 2*wv['woman']
wv.most_similar(positive = new_vector2)
```

Even if you didn't miss it, someone in your class probably did. Why? The word was right there, and you read it. So, why didn't your brain process it and notice the error? Because **you weren't paying attention** to that part of the text. I gave you a task, find the error in the numbers, so your brain likely focused your attention on the numbers, deciding the exact text was less important. Similarly, the word "with" isn't super critical to understanding the meaning of the above paragraph, so your brain didn't pay much attention to it. This skill, the ability to find what is important and pay attention to it while paying less mind to the rest, is what the attention mechanism in neural networks seeks to accomplsih.

Why do we need an attention mechanism? Well, we discussed earlier about a major shortcoming with sequence-to-sequence models: they struggle to capture long-range dependencies. NLP models would read something at the beginning of the sentence, but completely forget it by the time they got to the end of the sentence. Consequently, NLP model performance dropped as the input sequences got longer and longer. Analyzing large swaths of text was difficult, and likely to misinterpret the text. The attention mechanism compensates for this by learning a weight between words in sentences -- it learns what words in a sequence matter to other words, and which words are less important.

This relatively simple idea is the key behind the most prominent modern NLP architecture: the transformer.

##Transformers: Not the Autobot kind

Even though we don't have any transforming tractor trailor trucks in store for you here, the neural network transformer architecture is still pretty interesting.

Transformer models work in two steps: encoding and decoding. Take a look at the below illustration.


image.png

This is an example of how transformer is applied to a translation task. First, you take a sentence in your input language and pass it into a stack of encoders. The output of the encoding layers is a word embedding (like we covered earlier when discussing word2vec) that represents the input sentence. The decoders work in opposite direction -- they take the word embedding and transform it into text. See below for a illustration of the encoder architectecture.



image.png

As we mentioned previously, the key innovation of the transformer architecture is the attention layer (here called self-attention). When we apply an attention layer, the word embedding becomes a *context embedding*. We call it a context embedding because we are no longer just encoding the word itself (like we did with word2vec). Instead, the attention layer encodes information from the important contextual words within the sequence into each word representation. Because transformer word embeddings encode sequence context, we cannot do math directly in the same way we could do with word embeddings. However, the contextual information included in transformer word embeddings is incredibly useful to a plethora of tasks.

The decoder layer is largely the same, but it has a slightly different attention mechanism. The specific ways in which it is different isn't super important. The primary difference between encoder layers and decoder layers is that encoders map from text to word embedding and decoders map from word embedding to text. #How do we do NLP?

So, we have covered the history of natural language processing, the fundamentals of modern NLP, and we have practiced using word2vec, which was foundational to the transformer architecture. Now, we will explain the *how* of natural language processing. The general flow of NLP tasks can be broken up into three main stages: tokenizaton, word embedding, and down-stream task. The word embedding stage is what we have covered thus far -- you use some model, such as a transformer, to transform the input sentence into a word embedding.

## Training

But, how do we train these large transformer models? Up to this point, all of the neural network applications that we have dealt with have been *supervised*. That means that we provided a labeled dataset for the model to learn on, so it knew what the "right answer" was supposed to be when it calculated the loss. However, we have been talking about our model learning a generalized representation of language -- not training it on one specific task for which there is a clear right answer. Additionally, the datasets we use to train these large language model transformers are *massive*. For example, BERT (the model that drives google's search engine) was trained on the *entirety* of Wikipedia *and every unpublished book Google could get their hands on*. Imagine trying to hire a human to label all that data in any way. That is a *monumental* task that would take lifetimes to perform.

Obviously, manually labelling the dataset is a nonstarter. Now, it's possible that Google could have developed some clever crowdsourcing approach where they had their users label their own data somehow, but that comes with it's own problems. Instead, large language models are trained using an *unsuprevised* training scheme. Unsuprevised training schemes create a training label using the nature of the data itself.

For example, one task on which BERT is trained is the "Masked Language Modeling Task". In this task, a word or two in each sentence is randomly selected and hid from the model. Then, they ask the model to guess what the hidden (or masked) word is, based on what the surrounding words are. Using approaches like this allows the model to learn from all the text (the corpus, as it is called in NLP) without much human intervention. After the language model has learned a general representation of language from these very large copora, we can fine tune them on our specific datasets.

## Fine-Tuning

When we fine tune a model, we use a small amount of data to make the general language model more specific to our task. For example, let's say you wanted to build a chatbot to talk to patients in healthcare settings -- perhaps to give them information on perscriptions. Our general language model has undoubetdly learned from medical texts which would have exposed it to medical-domain language. But, to make it perform even better, we can fine-tune it on specific texts related to conversations in this area. Perhaps, we could record conversations between pharmacists and patients at a local pharmacy. Training on this corpus would allow our language model to become familair with the specific langauge used at that particular pharmacy -- it can learn to speak like a pharmacist. When we use large language models, fine-tuning in this manner allows models to learn the nuances of specific scenarios, without having to learn from scratch. In other words, we use transfer learning -- we train the model on a large amount of text that isn't directly relevant to our use case so that it can build the necessary foundation, then we fine-tune it on our use case.

Fine Tuning in Action: Sentiment Analysis

Now, we are going to walk through what the fine tuning process looks like. We are going to use the IMDB dataset -- a dataset of movie reviews -- to perform sentiment analysis. Let's see what happens when we load the model up and perform sentiment analysis on some example text.

```python
from transformers import BertForSequenceClassification, BertTokenizerFast, Trainer, TrainingArguments, pipeline
import torch
from datasets import load_dataset
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
```

```python
model = BertForSequenceClassification.from_pretrained('bert-base-uncased').to('cuda')
tokenizer = BertTokenizerFast.from_pretrained('bert-base-uncased')
pipe = pipeline("text-classification", model = model, tokenizer = tokenizer,  device=0)

print(pipe("I thought this movie sucks"))
print(pipe("I thought this movie is great"))
```

Hmm, it seems like our model currently doesn't see much of a difference between our two inputs. Let's see what happens when we fine tune our model on part of the IMDB dataset. This code will probably take a while to run, so in the meantime, if you have never tried it out, why don't you experiment with ChatGPT?
```python
model2 = BertForSequenceClassification.from_pretrained('bert-base-uncased')
model2.to('cuda')
tokenizer2  = BertTokenizerFast.from_pretrained('bert-base-uncased')

def tokenize(batch):
    return tokenizer2(batch['text'], padding=True, truncation=True)

train_dataset, test_dataset = load_dataset('imdb', split=['train[25%:75%]', 'test[25%:75%]'])

train_dataset = train_dataset.map(tokenize, batched=True, batch_size=len(train_dataset))
test_dataset = test_dataset.map(tokenize, batched=True, batch_size=len(train_dataset))
train_dataset.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])
test_dataset.set_format('torch', columns=['input_ids', 'attention_mask', 'label'])


def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
    acc = accuracy_score(labels, preds)
    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=1,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=64,
    warmup_steps=500,
    weight_decay=0.01,

)

trainer = Trainer(
    model=model2,
    args=training_args,
    #compute_metrics=compute_metrics,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

trainer.train()

trainer.evaluate()
```

```python
pipe2 = pipeline("text-classification", model = model2, tokenizer = tokenizer2, device = 0)
```

```python
print(pipe2("This movie is really awful"))
print(pipe2("I loved this movie"))
```

Now, we see the model is able to classify our text with a much greater level of accuracy. Even though the model had been trained before by google, it needed to learn our specific task. By giving it just a few minutes to pour over the sentiments in our dataset, the model learned our specific task.
